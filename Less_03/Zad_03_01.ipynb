```
import json
import numpy as np
import pandas as pd
import re
```

```
records = []
with open("/content/gazeta_test.jsonl", "r") as r:
    for line in r:
        records.append(json.loads(line))
```

```
df_test = pd.DataFrame(columns=('teg','text'))
```

```
for i in range(len(records)):
  fg = records[i].get('url')
  result = re.sub('/', '', re.findall(r'/\w+', fg)[1])
  df_test.loc[i, 'teg'] = result
  fg1 = records[i].get('text')
  df_test.loc[i, 'text'] = fg1
  

df_test
```
